```python
#char level 

#encode character level from https://medium.com/@bramblexu/character-level-cnn-with-keras-50391c3adf33


tk = Tokenizer(num_words=None, char_level=True, oov_token='UNK')
tk.fit_on_texts(data['text'])
# If we already have a character list, then replace the tk.word_index
# If not, just skip below part

# -----------------------Skip part start--------------------------
# construct a new vocabulary
alphabet = "abcdefghijklmnopqrstuvwxyz0123456789,;.!?:'\"/\\|_@#$%^&*~`+-=<>()[]{}"
char_dict = {}
for i, char in enumerate(alphabet):
    char_dict[char] = i + 1

# Use char_dict to replace the tk.word_index
tk.word_index = char_dict.copy()
# Add 'UNK' to the vocabulary
tk.word_index[tk.oov_token] = max(char_dict.values()) + 1
# -----------------------Skip part end----------------------------

# Convert string to index
charSeq = tk.texts_to_sequences(data['text'])
charSeq = pad_sequences(charSeq, maxlen = 280)

charEmbed = []
charEmbed.append(np.zeros(len(tk.word_index)))

for char, i in tk.word_index.items():
    embedding = np.zeros(len(tk.word_index))
    embedding[i-1] = 1
    charEmbed.append(embedding)
charEmbed= np.array(charEmbed)
#create Model

charInput = Input(shape=(280,))
charEmbedding=Embedding(len(tk.word_index) + 1,
                            128,
                            weights=[charEmbed],
                            input_length=280,
                            ) (charInput)

charConv = Conv1D(filters=512, kernel_size = 5, activation = 'relu')(charEmbedding)
charFlat = Flatten()(charConv)
charModel=Dense(128, activation = 'relu')(charFlat)
```

```python
from keras.models import Sequential
from keras.layers import Conv1D
from keras.layers import Embedding
from keras.layers import Dense
from keras.layers import Activation
from keras.layers import Flatten
from keras.layers import MaxPooling1D
from keras.layers import Dropout
from keras.layers import concatenate
from keras.models import Model
from keras.layers import Input

seqInput = Input(shape = (33,))
concept=Sequential()
concept=Embedding(len(word_index) + 1,
                            conceptDimension,
                            weights=[concept_matrix],
                            input_length=33,
                            )(seqInput)

wordvec = Sequential()
wordvec=Embedding(len(word_index) + 1,
                            100,
                            #weights=[embedding_matrix],
                            input_length=33,
                            )(seqInput)
modelInput = concatenate([wordvec, concept], axis = 0)
print(dir(modelInput))
model=Sequential()
model.add(Conv1D(filters=512, kernel_size = 5, input_shape = modelInput.get_shape()))
model.add(Activation('relu'))
model.add(Conv1D(filters=128, kernel_size = 5))
model.add(Activation('relu'))
model.add(Conv1D(filters=128, kernel_size = 5))
model.add(Activation('relu'))
model.add(Conv1D(filters=128, kernel_size = 5))
model.add(Activation('relu'))
model.add(Conv1D(filters=128, kernel_size = 5))
model.add(Activation('relu'))
model.add(Conv1D(filters=128, kernel_size = 5))
model.add(Activation('relu'))
model.add(Flatten())
model.add(Dense(128))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(1))
model.add(Activation('relu'))

cnn = Model(inputs=seqInput, ouputputs = model)

cnn.compile(optimizer='rmsprop',
              loss='mse')

history = cnn.fit(paddedseq, np.array(paddedtext['labels']),
                    batch_size=100,
                    epochs=5,
                    verbose=1,
                    validation_split=0.3)
```
